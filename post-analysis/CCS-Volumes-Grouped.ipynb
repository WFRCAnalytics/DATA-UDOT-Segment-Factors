```python
import pandas as pd
```

```python
# get data for station 634
dfCcsHourDirVol = pd.read_csv('../intermediate/CCSHourDir_Combined_2013to2019.csv')
dfCcsHourDirVol = dfCcsHourDirVol[(dfCcsHourDirVol['STATION']==634) & (dfCcsHourDirVol['YEAR']<=2019)]
dfCcsHourDirVol['DATETIME'] = pd.to_datetime(dfCcsHourDirVol[['YEAR', 'MONTH', 'DAY', 'HOUR']])

dfCcsHourDirVol
```

```python
dfCcsHourDirVol[(dfCcsHourDirVol['YEAR']==2019)]
```

```python
import plotly.express as px

# Assuming df is your DataFrame and it's structured with 'DATE', 'HOURVOL', and 'DIR' columns
# Replace df with the name of your DataFrame variable

# Creating the plot
fig = px.scatter(dfCcsHourDirVol, x='DATETIME', y='HOURVOL', color='DIR', title='Hourly Volume by Direction')

# Show the plot
fig.show()
```

```python
# Creating the plot
fig = px.scatter(dfCcsHourDirVol[dfCcsHourDirVol['YEAR']==2017], x='DATETIME', y='HOURVOL', color='DIR', title='Hourly Volume by Direction')

# Show the plot
fig.show()
```

```python
# Creating the plot
fig = px.scatter(dfCcsHourDirVol[(dfCcsHourDirVol['YEAR']==2017) & (dfCcsHourDirVol['MONTH']==10)], x='DATETIME', y='HOURVOL', color='DIR', title='Hourly Volume by Direction')

# Show the plot
fig.show()
```

```python
# Creating the scatter plot with lines
fig = px.line(dfCcsHourDirVol[(dfCcsHourDirVol['YEAR']==2017) & (dfCcsHourDirVol['MONTH']==10)], x='DATETIME', y='HOURVOL', color='DIR', title='Hourly Volume Scatter Plot with Lines by Direction - CCS #634',
                 markers='true')

# Show the plot
fig.show()
```

```python
# Filter your DataFrame as required
filtered_df = dfCcsHourDirVol[(dfCcsHourDirVol['YEAR']==2017) & (dfCcsHourDirVol['MONTH']==10)]

# Convert 'DATETIME' to date-only
filtered_df['DATE'] = filtered_df['DATETIME'].dt.date

# Group by the new 'DATE' column and 'DIR', and sum 'HOURVOL'
aggregated_df = filtered_df.groupby(['DATE', 'DIR'],as_index=False).agg(DAYVOL=('HOURVOL','sum'))

# Creating the scatter plot with lines
fig = px.line(aggregated_df, x='DATE', y='DAYVOL', color='DIR', 
              title='Daily Volume Scatter Plot with Lines by Direction - CCS #634',
              markers=True)

# Show the plot
fig.show()
```

```python
# Creating the scatter plot with lines
fig = px.line(dfCcsHourDirVol[(dfCcsHourDirVol['YEAR']==2017) & (dfCcsHourDirVol['MONTH']==10) & (dfCcsHourDirVol['DAY']==4) & (dfCcsHourDirVol['DOW']==2)], x='HOUR', y='HOURVOL', color='DIR', title='Hourly Volume Scatter Plot with Lines by Direction - CCS #634 - October 4, 2017',
                 markers='true')

# Show the plot
fig.show()
```
