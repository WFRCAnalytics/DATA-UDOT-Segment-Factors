```python
#Check for duplicate data
# Assuming your dataframe is named 'df'

# Define the key columns that identify duplicates
key_columns = ['STATION', 'DATE', 'ROUTE', 'MP', 'LANE']

# Define the hourly volume columns (all other columns except key_columns)
hourly_columns = [col for col in df_CCSHourDirLane.columns if col not in key_columns]

# Step 1: Create a helper column to check if a row has data in hourly volumes
df_CCSHourDirLane['has_hourly_data'] = df_CCSHourDirLane[hourly_columns].notna().any(axis=1)

# Step 2: Sort rows such that those with hourly data appear first within each group
df_CCSHourDirLane = df_CCSHourDirLane.sort_values(by=key_columns + ['has_hourly_data'], ascending=[True, True, True, True, True, False])

# Step 3: Drop duplicate rows based on the key columns, keeping the first occurrence
df_CCSHourDirLane = df_CCSHourDirLane.drop_duplicates(subset=key_columns, keep='first')

# Step 4: Drop the helper column
df_CCSHourDirLane = df_CCSHourDirLane.drop(columns=['has_hourly_data'])

# Display the cleaned dataframe
display(df_CCSHourDirLane)
```

```python
df_CCSHourDirLane = df_CCSHourDirLane.drop(columns=['ROUTE', 'MP'])

#normalize dataset to make calculations easier
dColHours = ['0000','0100','0200','0300','0400','0500','0600','0700','0800','0900','1000','1100','1200','1300','1400','1500','1600','1700','1800','1900','2000','2100','2200','2300']
df_CCSHourDirLane = pd.melt(df_CCSHourDirLane, id_vars=['STATION','DATE','LANE'], value_vars=dColHours)

#rename melted columns
df_CCSHourDirLane = df_CCSHourDirLane.rename(columns={"variable": "HOUR", "value": "HOURVOL"})

#drop rows with null volumes
df_CCSHourDirLane = df_CCSHourDirLane.dropna()

#convert string hour to integer
df_CCSHourDirLane['HOUR'] = df_CCSHourDirLane['HOUR'].str[-2:]
df_CCSHourDirLane['HOUR'] = df_CCSHourDirLane['HOUR'].astype('int')

#convert volume to integer
df_CCSHourDirLane['HOURVOL'] = df_CCSHourDirLane['HOURVOL'].astype('int')

#split LANE values into direction and lane number
df_CCSHourDirLane['DIR' ] = df_CCSHourDirLane['LANE'].str[0]
df_CCSHourDirLane['LANE'] = df_CCSHourDirLane['LANE'].str[5].astype(int)

#get rid of negative sign on station ids
df_CCSHourDirLane['STATION'] = df_CCSHourDirLane['STATION'] * -1

#set month field
df_CCSHourDirLane['YEAR' ] = df_CCSHourDirLane['DATE'].dt.year
df_CCSHourDirLane['MONTH'] = df_CCSHourDirLane['DATE'].dt.month
df_CCSHourDirLane['DAY'  ] = df_CCSHourDirLane['DATE'].dt.day
df_CCSHourDirLane['DOW'  ] = df_CCSHourDirLane['DATE'].dt.dayofweek

#df_CCSHourDirLane['DOW'] = df_CCSHourDirLane.apply(lambda row: daynum(row['DOWNAME']),axis=1)

df_CCSHourDirLane = df_CCSHourDirLane[['STATION','DIR','YEAR','MONTH','DAY','LANE','HOUR','HOURVOL','DOW']]

display(df_CCSHourDirLane)
```

```python
df_CCSHourDir = df_CCSHourDirLane.groupby(['STATION','YEAR','MONTH','DAY','DOW','HOUR','DIR'],as_index=False).agg(HOURVOL=('HOURVOL','sum'))
df_CCSHourDir
```

```python
#get hourly average for each month for each hour

#set month field
df_CCSAvg_MonthDOWHourDir = df_CCSHourDir.groupby(['STATION','DIR','YEAR','MONTH','DOW','HOUR'],as_index=False).agg(AVGHOURVOL=('HOURVOL','mean'))
df_CCSAvg_MonthDOWHourDir
```

```python
#for checks and understanding
#calculate number of days with data per station per month
df_1 = df_CCSHourDir.groupby(['STATION','MONTH','YEAR','DAY'],as_index=False).agg(NUMHOURS=('HOURVOL','size'))
df_2 = df_1.groupby(['STATION','YEAR','MONTH'],as_index=False).agg(NUMDAYS=('DAY','size'))
df_3 = df_2.pivot_table(index=['STATION'], columns=('YEAR','MONTH'), values='NUMDAYS').reset_index()

df_3 = df_3.fillna(0)
df_3 = df_3.astype(np.int32)

#export
df_3.to_csv(os.path.join(dirIntermediate, r'StationMonthDays.csv'),index=False)
```
