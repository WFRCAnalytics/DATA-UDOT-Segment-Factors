```python
import numpy as np
import os
import pandas as pd
import arcpy

# Import the required ArcGIS API for Python modules
import arcgis
from arcgis.gis import GIS
gis = arcgis.GIS()
from arcgis.geoanalytics import manage_data

from IPython.display import display, Markdown
```

```python
dirWork         = os.getcwd()
dirInput        = os.path.join(dirWork, r'input'       )
dirIntermediate = os.path.join(dirWork, r'intermediate')
dirResults      = os.path.join(dirWork, r'results'     )
dirConfigs      = os.path.join(dirWork, r'configs'     )
```

```python
#import excel spreadsheet

filename_File1 = os.path.join(dirInput, "udot\\HourlyCCSData\\2018 Station 301-431.xlsx")
filename_File2 = os.path.join(dirInput, "udot\\HourlyCCSData\\2018 Station 501-716.xlsx")
filename_File3 = os.path.join(dirInput, "udot\\HourlyCCSData\\2019 Station 301-513.xlsx")
filename_File4 = os.path.join(dirInput, "udot\\HourlyCCSData\\2019 Station 601-716.xlsx")

xl_file1 = pd.ExcelFile(filename_File1)
xl_file2 = pd.ExcelFile(filename_File2)
xl_file3 = pd.ExcelFile(filename_File3)
xl_file4 = pd.ExcelFile(filename_File4)

dfs1 = {sheet_name: xl_file1.parse(sheet_name) 
        for sheet_name in xl_file1.sheet_names}
dfs2 = {sheet_name: xl_file2.parse(sheet_name) 
        for sheet_name in xl_file2.sheet_names}
dfs3 = {sheet_name: xl_file3.parse(sheet_name) 
        for sheet_name in xl_file3.sheet_names}
dfs4 = {sheet_name: xl_file4.parse(sheet_name) 
        for sheet_name in xl_file4.sheet_names}
```

```python
df_CCSHourDirLane = dfs1['0301'][0:0] #just copy data structure, no row data

#take out of dictionary and put into single dataframe
for key in dfs1:
    #display(key)
    if (len(key)==4):
        dfs1[key].rename({c: c[1:] for c in dfs1[key].columns if c.startswith('H')}, axis='columns', inplace=True)
        df_CCSHourDirLane = df_CCSHourDirLane.append(dfs1[key], ignore_index=True)
for key in dfs2:
    #display(key)
    if (len(key)==4):
        dfs2[key].rename({c: c[1:] for c in dfs2[key].columns if c.startswith('H')}, axis='columns', inplace=True)
        df_CCSHourDirLane = df_CCSHourDirLane.append(dfs2[key], ignore_index=True)
for key in dfs3:
    #display(key)
    if (len(key)==4):
        dfs3[key].rename({c: c[1:] for c in dfs3[key].columns if c.startswith('H')}, axis='columns', inplace=True)
        df_CCSHourDirLane = df_CCSHourDirLane.append(dfs3[key], ignore_index=True)
for key in dfs4:
    #display(key)
    if (len(key)==4):
        dfs4[key].rename({c: c[1:] for c in dfs4[key].columns if c.startswith('H')}, axis='columns', inplace=True)
        df_CCSHourDirLane = df_CCSHourDirLane.append(dfs4[key], ignore_index=True)
        
df_CCSHourDirLane
```

```python
df_CCSHourDirLane = df_CCSHourDirLane.drop(columns=['ROUTE', 'MP'])

#normalize dataset to make calculations easier
dColHours = ['0000','0100','0200','0300','0400','0500','0600','0700','0800','0900','1000','1100','1200','1300','1400','1500','1600','1700','1800','1900','2000','2100','2200','2300']
df_CCSHourDirLane = pd.melt(df_CCSHourDirLane, id_vars=['STATION','DATE','LANE'], value_vars=dColHours)

#rename melted columns
df_CCSHourDirLane = df_CCSHourDirLane.rename(columns={"variable": "HOUR", "value": "HOURVOL"})

#drop rows with null volumes
df_CCSHourDirLane = df_CCSHourDirLane.dropna()

#convert string hour to integer
df_CCSHourDirLane['HOUR'] = df_CCSHourDirLane['HOUR'].str[-2:]
df_CCSHourDirLane['HOUR'] = df_CCSHourDirLane['HOUR'].astype('int')

#convert volume to integer
df_CCSHourDirLane['HOURVOL'] = df_CCSHourDirLane['HOURVOL'].astype('int')

#split LANE values into direction and lane number
df_CCSHourDirLane['DIR' ] = df_CCSHourDirLane['LANE'].str[0]
df_CCSHourDirLane['LANE'] = df_CCSHourDirLane['LANE'].str[5].astype(int)

#get rid of negative sign on station ids
df_CCSHourDirLane['STATION'] = df_CCSHourDirLane['STATION'] * -1

#set month field
df_CCSHourDirLane['YEAR' ] = df_CCSHourDirLane['DATE'].dt.year
df_CCSHourDirLane['MONTH'] = df_CCSHourDirLane['DATE'].dt.month
df_CCSHourDirLane['DAY'  ] = df_CCSHourDirLane['DATE'].dt.day
df_CCSHourDirLane['DOW'  ] = df_CCSHourDirLane['DATE'].dt.dayofweek

#df_CCSHourDirLane['DOW'] = df_CCSHourDirLane.apply(lambda row: daynum(row['DOWNAME']),axis=1)

df_CCSHourDirLane = df_CCSHourDirLane[['STATION','DIR','YEAR','MONTH','DAY','LANE','HOUR','HOURVOL','DOW']]

display(df_CCSHourDirLane)
```

```python
df_CCSHourDir = df_CCSHourDirLane.groupby(['STATION','YEAR','MONTH','DAY','DOW','HOUR','DIR'],as_index=False).agg({'HOURVOL':[np.sum]})
df_CCSHourDir = df_CCSHourDir.droplevel(1, axis=1)
df_CCSHourDir
```

```python
df_CCSHourDir.dtypes
```

```python
#get hourly average for each month for each hour

#set month field
df_CCSAvg_MonthDOWHourDir = df_CCSHourDir.groupby(['STATION','DIR','YEAR','MONTH','DOW','HOUR'],as_index=False).agg({'HOURVOL':[np.mean]})
df_CCSAvg_MonthDOWHourDir = df_CCSAvg_MonthDOWHourDir.droplevel(1, axis=1)
df_CCSAvg_MonthDOWHourDir = df_CCSAvg_MonthDOWHourDir.rename(columns={"HOURVOL": "AVGHOURVOL"})
df_CCSAvg_MonthDOWHourDir
```

```python
#for checks and understanding
#calculate number of days with data per station per month
df_1 = df_CCSHourDir.groupby(['STATION','MONTH','YEAR','DAY'],as_index=False).agg({'HOURVOL':[np.size]})
df_2 = df_1.groupby(['STATION','YEAR','MONTH'],as_index=False).agg({'DAY':[np.size]})
df_2 = df_2.droplevel(1, axis=1)
df_3 = df_2.pivot_table(index=['STATION'], columns=('YEAR','MONTH'), values='DAY').reset_index()

df_3 = df_3.fillna(0)
df_3 = df_3.astype(np.int32)

#export
df_3.to_csv(os.path.join(dirIntermediate, r'StationMonthDays.csv'),index=False)
```
